```python
gemini_baseline = pd.read_csv("/Users/ashmi/Code/Scripts/phd/sustainable-TRS-RAG/sustainable-trs-with-rag/european-city-data/rag-sustainability/results/results-combined_prompts/llm-judge/judged_cities_Gemini1.5Pro_v2.csv")
print(gemini_baseline.shape)
gemini_baseline.rename(columns={"response_sustainable": "response", "llm_judge_response_sustainable": "llm_judge_response"}, inplace=True)
# gemini_baseline = gemini_baseline[["prompt_id","model", "prompt", "response", 'llm_judge_response', 'total_rating']]
# gemini_baseline.to_csv("/Users/ashmi/Code/Scripts/phd/sustainable-TRS-RAG/sustainable-trs-with-rag/european-city-data/rag-sustainability/results/results-combined_prompts/llm-judge/judged_cities_Gemini1.5Pro_v2.csv", index=False)
gemini_baseline['total_rating'] = gemini_baseline['llm_judge_response'].apply(lambda x: extract_judge_score(x))
gemini_baseline.head()
```

```python
gemini_sar = pd.read_csv("/Users/ashmi/Code/Scripts/phd/sustainable-TRS-RAG/sustainable-trs-with-rag/european-city-data/rag-sustainability/results/results-combined_prompts_SAR/llm-judge/judged_cities_Gemini1.5Pro_v2.csv")
gemini_sar.drop(columns=["response"], inplace=True)
print(gemini_sar.shape)
gemini_sar['total_rating_sar'] = gemini_sar['llm_judge_response_sustainable'].apply(lambda x: extract_judge_score(x))
gemini_sar.head()
```

```python
gemini = gemini_baseline.merge(gemini_sar, on=["prompt_id","model", "prompt"], how="left")
gemini.head()
```

```python
def compute_mean(df):
    llama_results = df.loc[df["model"]=="llama3point1-instruct"]
    print("llama shape:", llama_results.shape)
    
    mistral_results = df.loc[df["model"] == "mistral-instruct"]
    print("mistral shape:", mistral_results.shape)
    
    gemma_results = df.loc[df["model"] == "gemma2"]
    print("gemma shape:", gemma_results.shape)
    
    print("\t llama results mean (Baseline)", llama_results.total_rating.mean())
    print("\t llama results mean (Sustainable)", llama_results.total_rating_sar.mean())
    
    print("\t Mistral results mean (Baseline)", mistral_results.total_rating.mean())
    print("\t Mistral results mean (Sustainable)", mistral_results.total_rating_sar.mean())

    print("\t Gemma results mean (Baseline)", gemma_results.total_rating.mean())
    print("\t Gemma results mean (Sustainable)", gemma_results.total_rating_sar.mean())
    
```

```python
def compute_sd(df):
    llama_results = df.loc[df["model"]=="llama3point1-instruct"]
    mistral_results = df.loc[df["model"] == "mistral-instruct"]
    gemma_results = df.loc[df["model"] == "gemma2"]
    
    print("\t llama results SD (Baseline)", llama_results.total_rating.std())
    print("\t llama results SD (Sustainable)", llama_results.total_rating_sar.std())
    
    print("\t Mistral results SD (Baseline)", mistral_results.total_rating.std())
    print("\t Mistral results SD (Sustainable)", mistral_results.total_rating_sar.std())
    
    print("\t Gemma results SD (Baseline)", gemma_results.total_rating.std())
    print("\t Gemma results SD (Sustainable)", gemma_results.total_rating_sar.std())
```

```python
compute_mean(gemini)
```

```python
compute_sd(gemini)
```
