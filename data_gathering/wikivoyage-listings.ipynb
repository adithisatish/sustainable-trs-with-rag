```python
cities = pd.read_csv("../city_abstracts_embeddings.csv")
```

```python
eu_listings = listings[listings['article'].isin(cities['city'])]
```

```python
eu_listings = eu_listings[['article', 'type', 'title', 'description']]
```

```python
required = ['see', 'do', 'go', 'view', 'eat', 'drink', 'sleep']
eu_listings = eu_listings[eu_listings['type'].isin(required)]

eu_listings
```

```python
eu_listings['description'].fillna("No description found.", inplace=True)
```
