```python
import os
import pandas as pd
import numpy as np
```

```python
wikivoyage_dir = "rag-sustainable-trs/"
airbnb_dir = "green-city-reco-data/airbnb_data/calendar_data/"
city_trends_dir = "green-city-reco-data/trends_data/"
eu_cities_dir = ['europeancities-wikivoyage-tripadvisor/data-sources/wikivoyage/', 'europeancities-wikivoyage-tripadvisor/data-sources/tripadvisor/']
```

```python
city_abstracts = pd.read_csv(wikivoyage_dir + "city_abstracts_embeddings.csv")
city_trends = pd.read_csv(city_trends_dir + "city_trends_clean.csv")
```

```python
airbnb_cities = os.listdir(airbnb_dir)
len(airbnb_cities)

airbnb_cities = list(map(lambda x: (x.split("-")[0]).capitalize(), airbnb_cities))
airbnb_cities
```

```python
cities_wv = set(city_abstracts['city'])
cities_gt = set(city_trends['city'])
```

```python
result = cities_wv & cities_gt
len(result)
```

```python
result
```

```python
mapping_airbnb = pd.read_csv("green-city-reco-data/airbnb_data/airbnb_city_country_mapping.csv", sep=";")
mapping_airbnb.head()
```

```python
final_results = result & set(mapping_airbnb['city'])

len(final_results)
```

```python
# result == result_without_trends
countries = city_abstracts[city_abstracts['city'].isin(final_results)]
cities = countries[['city', 'country']]
cities.reset_index(inplace=True)
cities.drop(columns='index', inplace=True)
cities.to_csv('europeancities.csv')
# for i in final_results: 
#     country = city_abstracts[city_abstracts['city'] == i]['country'].item()
#     countries.append(country)
```
