```python
import pandas as pd 
import numpy as np 
import requests
from bs4 import BeautifulSoup
import xml.etree.ElementTree as ET 
```

```python
with open("enwikivoyage-latest-pages-articles-multistream.xml") as file:
    tree = ET.parse(file)
```

```python
titles = tree.findall(".//title")
for title in titles:
    print(title.text)

print(len(titles))
```

```python
city_emb = pd.read_csv("city_abstracts_embeddings.csv")
city_emb.head()

list_of_cities = list(city_emb['city'])
list_of_cities
```

```python
pages = tree.findall(".//page")
# print(len(pages))
relevant_page_ids = []

document_path = "wikivoyage_city_documents/"

for page in pages:
    page_title = page.find("title").text
    # print(page_title)
    if page_title in list_of_cities:
        # print("yes")
        id = page.find("id").text
        relevant_page_ids.append(id)
        # print([child.tag for child in page])
        revision = page.find("revision")
        doc = revision.find("text").text

        with open(document_path + id + "_" + page_title + ".txt", "w+") as file:
            file.write(doc)

    # for children in page.iter(): 
    #     for child in children:
    #         print(child)
    #         # if child.attrib['title'] in list_of_cities:
    #         #     relevant_page_ids.append(child.attrib["id"])
```
