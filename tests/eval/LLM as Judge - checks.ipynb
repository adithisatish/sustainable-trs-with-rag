```python
import pandas as pd
from sklearn.metrics import mean_absolute_percentage_error as mape
```

```python
claude_new = pd.read_csv("../../../sustainable-trs-with-rag/european-city-data/rag-sustainability/results/results-combined_prompts_SAR/llm-judge/judged_cities_Claude3Point5Sonnet.csv")
claude_new.drop(columns=["response", 'llm_judge_response', 'total_rating'], inplace=True)
```

```python
claude_old = pd.read_csv("../../../sustainable-trs-with-rag/european-city-data/rag-sustainability/results/results-combined_prompts/llm-judge/judged_cities_Claude3Point5Sonnet.csv")
claude_old = claude_old[["prompt_id","model", "prompt", "response", 'llm_judge_response', 'total_rating']]

claude_old.head()
```

```python
claude = claude_old.merge(claude_new, on=["prompt_id","model", "prompt"], how="left")
claude.head()
```

```python
gpt_new = pd.read_csv("../../../sustainable-trs-with-rag/european-city-data/rag-sustainability/results/results-combined_prompts_SAR/llm-judge/judged_cities_GPT-4.csv")
gpt_new.drop(columns=["response", 'llm_judge_response', 'total_rating'], inplace=True)
gpt_new.head()
```

```python
gpt_old = pd.read_csv("../../../sustainable-trs-with-rag/european-city-data/rag-sustainability/results/results-combined_prompts/llm-judge/judged_cities_GPT-4.csv")
gpt_old = gpt_old[["prompt_id","model", "prompt", "response", 'llm_judge_response', 'total_rating']]
gpt_old.head()
```

```python
gpt = gpt_old.merge(gpt_new, on=["prompt_id","model", "prompt"], how="left")
gpt.head()
```

```python
gpt.to_csv("../../../sustainable-trs-with-rag/european-city-data/rag-sustainability/results/results-combined_prompts_SAR/llm-judge/judged_cities_GPT-4.csv", index=False)
claude.to_csv("../../../sustainable-trs-with-rag/european-city-data/rag-sustainability/results/results-combined_prompts_SAR/llm-judge/judged_cities_Claude3Point5Sonnet.csv", index=False)
```

```python
def compute_mean(df):
    llama_results = df.loc[df["model"]=="llama3point1-instruct"]
    mistral_results = df.loc[df["model"] == "mistral-instruct"]
    
    print("\t llama results mean (Baseline)", llama_results.total_rating.mean())
    print("\t llama results mean (Sustainable)", llama_results.total_rating_sustainable.mean())
    
    print("\t Mistral results mean (Baseline)", mistral_results.total_rating.mean())
    print("\t Mistral results mean (Sustainable)", mistral_results.total_rating_sustainable.mean())
    
    
    
```

```python
def compute_sd(df):
    llama_results = df.loc[df["model"]=="llama3point1-instruct"]
    mistral_results = df.loc[df["model"] == "mistral-instruct"]
    
    print("\t llama results SD (Baseline)", llama_results.total_rating.std())
    print("\t llama results SD (Sustainable)", llama_results.total_rating_sustainable.std())
    
    print("\t Mistral results SD (Baseline)", mistral_results.total_rating.std())
    print("\t Mistral results SD (Sustainable)", mistral_results.total_rating_sustainable.std())
    
   
```

```python
print("Mean")
print("GPT-4 as a judge")
compute_mean(gpt)

print("Claude-3.5-sonnet as a judge")
compute_mean(claude)
```

```python
print("SD")
print("GPT-4 as a judge")
compute_sd(gpt)

print("Claude-3.5-sonnet as a judge")
compute_sd(claude)
```

```python
def compute_mape(combined):
    llama = combined.loc[combined["model"]=="llama3point1-instruct"]
    mistral = combined.loc[combined["model"]=="mistral-instruct"]
    mape_llama = mape(llama.total_rating_gpt4, llama.total_rating_claude)
    mape_llama_sustainable = mape(llama.total_rating_sustainable_gpt4, llama.total_rating_sustainable_claude)
    
    mape_mistral = mape(mistral.total_rating_gpt4, mistral.total_rating_claude)
    mape_mistral_sustainable = mape(mistral.total_rating_sustainable_gpt4, mistral.total_rating_sustainable_claude)
    
    print("\t LLama MAPE (Non Sustainable)", mape_llama)
    print("\t LLama MAPE (Sustainable)", mape_llama_sustainable)
    
    print("\n\t Mistral MAPE (Non Sustainable)", mape_mistral)
    print("\t Mistral MAPE (Sustainable)", mape_mistral_sustainable)
```

```python
columns = ["prompt_id", "model", "total_rating", "total_rating_sustainable"]

try:
    gpt = gpt[columns]
    claude = claude[columns]
except KeyError:
    gpt = gpt
    claude = claude

gpt.rename(columns={"total_rating": "total_rating_gpt4","total_rating_sustainable":"total_rating_sustainable_gpt4" }, inplace=True)
claude.rename(columns={"total_rating": "total_rating_claude","total_rating_sustainable":"total_rating_sustainable_claude" }, inplace=True)

combined = pd.merge(gpt, claude, on=["prompt_id", "model"], how="left")
combined.fillna(0, inplace=True)
combined
```

```python
compute_mape(combined)
```
