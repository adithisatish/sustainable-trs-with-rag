## Judge: Gemini1.5Pro

```python
gemini_baseline = pd.read_csv("../../../sustainable-trs-with-rag/european-city-data/rag-sustainability/results/results-combined_prompts/llm-judge/judged_cities_Gemini1.5Pro_v2.csv")
print(gemini_baseline.shape)
gemini_baseline = gemini_baseline[["prompt_id","model", "prompt", "response", 'llm_judge_response', 'total_rating']]
gemini_baseline.head()
```

```python
gemini_sar = pd.read_csv("../../../sustainable-trs-with-rag/european-city-data/rag-sustainability/results/results-combined_prompts_SAR/llm-judge/judged_cities_Gemini1.5Pro_v2.csv")
gemini_sar.drop(columns=["response"], inplace=True)
print(gemini_sar.shape)
gemini_sar.head()
```

```python
gemini = gemini_baseline.merge(gemini_sar, on=["prompt_id","model", "prompt"], how="left")
gemini.head()
```

```python
def compute_mean(df):
    llama_results = df.loc[df["model"]=="llama3point1-instruct"]
    # llama_results.dropna(inplace=True)
    print("llama shape:", llama_results.shape)
    
    mistral_results = df.loc[df["model"] == "mistral-instruct"]
    # mistral_results.dropna(inplace=True)
    print("mistral shape:", mistral_results.shape)
    
    gemma_results = df.loc[df["model"] == "gemma2"]
    # gemma_results.dropna(inplace=True)
    print("gemma shape:", gemma_results.shape)
    
    print("\t llama results mean (Baseline)", llama_results.total_rating.mean())
    print("\t llama results mean (Sustainable)", llama_results.total_rating_sustainable.mean())
    
    print("\t Mistral results mean (Baseline)", mistral_results.total_rating.mean())
    print("\t Mistral results mean (Sustainable)", mistral_results.total_rating_sustainable.mean())

    print("\t Gemma results mean (Baseline)", gemma_results.total_rating.mean())
    print("\t Gemma results mean (Sustainable)", gemma_results.total_rating_sustainable.mean())
    
    
    
```

```python
def compute_sd(df):
    llama_results = df.loc[df["model"]=="llama3point1-instruct"]
    mistral_results = df.loc[df["model"] == "mistral-instruct"]
    gemma_results = df.loc[df["model"] == "gemma2"]
    
    print("\t llama results SD (Baseline)", llama_results.total_rating.std())
    print("\t llama results SD (Sustainable)", llama_results.total_rating_sustainable.std())
    
    print("\t Mistral results SD (Baseline)", mistral_results.total_rating.std())
    print("\t Mistral results SD (Sustainable)", mistral_results.total_rating_sustainable.std())
    
    print("\t Gemma results SD (Baseline)", gemma_results.total_rating.std())
    print("\t Gemma results SD (Sustainable)", gemma_results.total_rating_sustainable.std())
```

```python
print("Mean")
print("GPT-4 as a judge")
compute_mean(gpt)

print("Claude-3.5-sonnet as a judge")
# compute_mean(claude)

print("Gemini-1.5-Pro as a judge")
compute_mean(gemini)
```

```python
print("SD")
print("GPT-4 as a judge")
compute_sd(gpt)

print("Claude-3.5-sonnet as a judge")
# compute_sd(claude)

print("Gemini-1.5-Pro as a judge")
compute_sd(gemini)
```
